Получили уравнение для фугоидной траеткории:

$$\begin{equation}
z(t)'' + \frac{g \,z(t)}{z_t}=g
\end{equation}$$

Переписали его в виде системы линейных дифф. уравнений:

$$\begin{eqnarray}
z'(t) &=& v(t)\\
v'(t) &=& g\left(1-\frac{z(t)}{z_t}\right)
\end{eqnarray}$$

Ввели новое обозначение для вектора неизвестных:

$$\begin{equation}
\vec{u}  = \begin{pmatrix} z \\ v \end{pmatrix}
\end{equation}$$

И переписали сисетму уранений в вектрной форме:

\begin{equation}
\vec{u}'(t)  = \begin{pmatrix} v \\ g-g\frac{z(t)}{z_t} \end{pmatrix}
\end{equation}

## Метод Эйлера

Обозначим  $u_n$ аппроксимацию решения на временном шаге $t_n$. Для получения численного решения системы уравненй нужно произвести последовательность вычислений по формуле:

$$u_{n+1} = u_n + \Delta t \,f(u_n).$$

В этом и состоит **Метод Эйлера**.

В случае уравнений фугоидных колебаний метод Эйлера дает следующий алгоритм:

\begin{align}
z_{n+1} & = z_n + \Delta t \, v_n \\
v_{n+1} & = v_n + \Delta t \left(g - \frac{g}{z_t} \, z_n \right).
\end{align}

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
T = 100
dt = 0.01
N = int(T/dt)+1
t = numpy.arange(0,T+dt,dt)
print t[-1]
```

## Постановка начальных условий

```python
z0 = 100. #altitude
v = 10 #upward velocity - pertrubation
zt = 100
g = 9.81

u = numpy.array([z0,v])

z = numpy.zeros(N)
z[0] = z0
```

```python
def euler_method(u,z):
    """Euler method for system of 2 equastions describing fugoid oscillations
    
    Parameters
    ----------
    z : array of float
        numerical solution.
    u : 2d vector
        initial condition.
    
        
    Returns
    -------
    z : array of float
        numerical solution.
    """
    for n in range(1,len(z)):
        u = u + dt*numpy.array([u[1],g*(1-u[0]/zt)])
        z[n] = u[0]
    return z
```

```python
u = numpy.array([z0,v])

z = numpy.zeros(N)
z[0] = z0
z = euler_method(u,z)
```

## Точное решение

Уравнение для нармонических колебаний, значит решение -- синус и косинус: 
$$
\begin{equation}
z(t) = A \sin \left(\sqrt{\frac{g}{z_t}} t \right) + B \cos \left(\sqrt{\frac{g}{z_t}} t \right) + z_t
\end{equation}
$$

Константы определяем из начальных условий:
$$\begin{eqnarray}
z(0) = z_0 \\
v(0) = v_0
\end{eqnarray}$$

После непродолжительных вычислений получим:

$$\begin{equation}
z(t) = v_0 \sqrt{\frac{z_t}{g}} \sin \left(\sqrt{\frac{g}{z_t}} t \right) + (z_0-z_t) \cos \left(\sqrt{\frac{g}{z_t}} t \right) + z_t
\end{equation}$$

## Рисуем точное решение

На одном графике нарисуем точное и  численное решение и сравним. Сначала нужно пересчитать численное решение

```python
v=10
u = numpy.array([z0,v])

z = numpy.zeros(N)
z[0] = z0
z = euler_method(u,z)
z_exact = v*(zt/g)**.5*numpy.sin((g/zt)**.5*t)+\
            (z0-zt)*numpy.cos((g/zt)**.5*t)+zt
```

```python
plt.figure(figsize=(10,4))
plt.ylim(40,160)             #y-axis plot limits
plt.tick_params(axis='both', labelsize=14) #increase font size for ticks
plt.xlabel('t', fontsize=14) #x label
plt.ylabel('z', fontsize=14) #y label
plt.plot(t,z)
plt.plot(t, z_exact)
plt.legend([u'Численное решение',u'Аналитическое решение']);
```

На первый взгляд аналитическое и численное решение совпадают хорошо, однако, можно заметить, что с ростом $t$ расхождение увеличивается.
Точность метода. Если при уменьшениии величины шага $\Delta t$ численное решение (аппроксимация) приближается к точному, то метод обладает свойством *сходимости*. Если при $\Delta t\rightarrow 0$ ошибка аппроксимации стремится к нулю, то метод обладает свойством *согласованности*.

## Сходимость

Для корректного сравнения решений необходимо использовать какую-нибудь __норму__, например, норму $L_1$:

$$ E = \Delta t \sum_{n=0}^N \left|z(t_n) - z_n\right| $$ -

что есть сумма разностей точного и численного решения в каждом узле сетки (с нормировочным множителем).
Проверяем наш метод с разными $\Delta t$ (`dt`)

### Выисляем ошибку аппроксимации

```python
def get_error(z, dt):
    """Returns the error relative to analytical solution using L-1 norm.
    
    Parameters
    ----------
    z : array of float
        numerical solution.
    dt : float
        time increment.
        
    Returns
    -------
    err : float
        L_{1} norm of the error with respect to the exact solution.
    """
    N = len(z)
    t = numpy.linspace(0.0, T, N)
    
    z_exact = v*(zt/g)**.5*numpy.sin((g/zt)**.5*t)+\
                (z0-zt)*numpy.cos((g/zt)**.5*t)+zt
    
    return dt * numpy.sum(numpy.abs(z-z_exact))
```

```python
# массив шагов по времени
dt_values = numpy.array([0.1, 0.05, 0.01, 0.005, 0.001, 0.0001])

# массив решений 
z_values = numpy.empty_like(dt_values, dtype=numpy.ndarray)
error_values = numpy.empty_like(dt_values)

for i, dt in enumerate(dt_values):
    N = int(T/dt)+1    # число шагов по времени меняется с уменьшением самого шага
    ### discretize the time using numpy.linspace() ###
    #t = numpy.linspace(0.0, T, N)

    # initial conditions
    u = numpy.array([z0, v])
    z = numpy.zeros(N)
    z[0] = z0
    
    # time loop - Euler method
    z = euler_method(u,z)
    
    z_values[i] = z.copy()    # store the total elevation calculation grid i
    error_values[i] = get_error(z_values[i], dt) ### call the function get_error() 
```

```python
plt.figure(figsize=(10, 6))
plt.tick_params(axis='both', labelsize=14) #increase tick font size
plt.grid(True)                         #turn on grid lines
plt.xlabel('$\Delta t$', fontsize=16)  #x label
plt.ylabel('Error', fontsize=16)       #y label
plt.loglog(dt_values, error_values, 'ko-')  #log-log plot
plt.axis('equal')                      #make axes scale equally;
```
